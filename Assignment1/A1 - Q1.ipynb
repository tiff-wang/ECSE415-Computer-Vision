```python
import cv2
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
peppers = cv2.imread('./peppers.png')
filter_size = 7
```

```python
noisy_peppers = cv2.imread('./gaussian_noise_peppers.png')
denoised_gaussian_peppers = cv2.GaussianBlur(noisy_peppers, (filter_size, filter_size), 2)

# plot
plt.subplot(121), plt.imshow(noisy_peppers)
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(denoised_gaussian_peppers)
plt.title("Gaussian Filter"), plt.xticks([]), plt.yticks([])
```

```python
noisy_peppers = cv2.imread('./gaussian_noise_peppers.png')
box_filter = np.ones((filter_size, filter_size), np.float32)/(filter_size * filter_size)
denoised_box_peppers = cv2.filter2D(noisy_peppers, -1, box_filter)

# plot
plt.subplot(121), plt.imshow(noisy_peppers)
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(denoised_box_peppers)
plt.title("Box Filter"), plt.xticks([]), plt.yticks([])
```

```python
mse_gaussian = np.mean( (denoised_gaussian_peppers - peppers) ** 2 )
mse_box = np.mean( (denoised_box_peppers - peppers) ** 2 )

R = 255
pnsr_gaussain = math.log10( R * R / math.sqrt(mse_gaussian) )
pnsr_box = math.log10( R * R / math.sqrt(mse_box) )

print("Gaussian Filter PSNR: ", pnsr_gaussain)
print("Box Filter PSNR: ", pnsr_box)
```

```python
noisy_saltAndPepper_peppers = cv2.imread('./saltandpepper_noise_peppers.png')
denoised_gaussian_peppers = cv2.GaussianBlur(noisy_saltAndPepper_peppers, (filter_size, filter_size), 2)

# plot
plt.subplot(121), plt.imshow(noisy_saltAndPepper_peppers)
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(denoised_gaussian_peppers)
plt.title("Gaussian Filter"), plt.xticks([]), plt.yticks([])
```

```python
noisy_saltAndPepper_peppers = cv2.imread('./saltandpepper_noise_peppers.png')
denoised_median_peppers = cv2.medianBlur(noisy_saltAndPepper_peppers, filter_size)

# plot
plt.subplot(121), plt.imshow(noisy_saltAndPepper_peppers)
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(denoised_median_peppers)
plt.title("Median Filter"), plt.xticks([]), plt.yticks([])
```

```python
mse_gaussian = np.mean( (denoised_gaussian_peppers - peppers) ** 2 )
mse_median = np.mean( (denoised_box_peppers - peppers) ** 2 )

R = 255
pnsr_gaussain = math.log10( R * R / math.sqrt(mse_gaussian) )
pnsr_median = math.log10( R * R / math.sqrt(mse_median) )

print("Gaussian Filter PSNR: ", pnsr_gaussain)
print("Median Filter PSNR: ", pnsr_median)
```

```python
rice = cv2.imread('./rice.png')
plt.imshow(rice)
```

```python
filter_size = 5
variance = 2
blurred_gaussian_rice = cv2.GaussianBlur(rice, (filter_size, filter_size), variance)
gaussian_detail = rice - blurred_gaussian_rice
sharpened_gaussian_rice = rice + gaussian_detail

box_filter = np.ones((filter_size, filter_size), np.float32)/(filter_size * filter_size)
blurred_box_rice = cv2.filter2D(rice, -1, box_filter)
box_detail = rice - blurred_box_rice
sharpened_box_rice = rice + box_detail

# plot
plt.subplot(121), plt.imshow(sharpened_gaussian_rice)
plt.title("Gaussian Sharpen"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sharpened_box_rice)
plt.title("Box Sharpen"), plt.xticks([]), plt.yticks([])
```
